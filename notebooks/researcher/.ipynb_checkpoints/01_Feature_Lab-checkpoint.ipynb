# 🔬 Feature Lab – Quant Researcher
**Workflow**: Load OHLCV → Compute features → Feature importance → Correlation with forward returns → Save to FeatureStore

---

```python
from qtrader.analyst import AnalystSession, RoleContext
from qtrader.features.store import FeatureStore
import polars as pl
import numpy as np
import matplotlib.pyplot as plt

session = AnalystSession(role=RoleContext.RESEARCHER)
session.info()

SYMBOL = 'BTC-USD'
TIMEFRAME = '1d'
```

## 1. Load & Engineer Features

```python
try:
    df = session.load_ohlcv(SYMBOL, TIMEFRAME)
except Exception:
    df = session.sample_ohlcv(symbol='BTC', days=365)

df = session.make_returns(df)
df = session.add_rolling_features(df, windows=[5, 10, 21, 63])
df = df.drop_nulls()

# Alpha scores (forward returns)
df = session.run_alpha_score(df, forward_periods=[1, 5, 10])
df = df.drop_nulls()
print(f"Feature-enriched shape: {df.shape}")
df.head(3)
```

## 2. Feature × Forward Return Correlation

```python
feature_cols = [c for c in df.columns if c.startswith(('sma_', 'vol_', 'rsi_', 'avg_'))]
target_cols  = ['fwd_ret_1', 'fwd_ret_5', 'fwd_ret_10']

corr_matrix = np.zeros((len(feature_cols), len(target_cols)))
for i, fc in enumerate(feature_cols):
    for j, tc in enumerate(target_cols):
        x, y = df[fc].to_numpy(), df[tc].to_numpy()
        corr_matrix[i, j] = float(np.corrcoef(x, y)[0, 1])

fig, ax = plt.subplots(figsize=(8, max(4, len(feature_cols)*0.5)), facecolor='#0f1117')
ax.set_facecolor('#0f1117')
im = ax.imshow(corr_matrix, cmap='RdYlGn', vmin=-0.3, vmax=0.3, aspect='auto')
ax.set_xticks(range(len(target_cols))); ax.set_xticklabels(target_cols, color='#e2e8f0', fontsize=10)
ax.set_yticks(range(len(feature_cols))); ax.set_yticklabels(feature_cols, color='#e2e8f0', fontsize=9)
plt.colorbar(im, ax=ax)
ax.set_title('Feature × Forward Return Correlation', color='#e2e8f0')
for i in range(len(feature_cols)):
    for j in range(len(target_cols)):
        ax.text(j, i, f'{corr_matrix[i,j]:.2f}', ha='center', va='center', color='white', fontsize=8)
plt.tight_layout()
plt.show()
```

## 3. Feature Drift Check (Rolling Correlation)

```python
# Check rolling 60-period correlation of top feature with fwd_ret_1
top_feat = feature_cols[np.argmax(np.abs(corr_matrix[:, 0]))]
print(f"Top feature for 1-period forward return: {top_feat}")

feat_arr = df[top_feat].to_numpy()
fwd_arr  = df['fwd_ret_1'].to_numpy()
win = 60
roll_corr = [np.corrcoef(feat_arr[i:i+win], fwd_arr[i:i+win])[0,1] for i in range(len(feat_arr)-win)]

fig2, ax = plt.subplots(figsize=(12, 3), facecolor='#0f1117')
ax.set_facecolor('#0f1117')
for sp in ax.spines.values(): sp.set_color('#334155')
ax.tick_params(colors='#94a3b8')
ax.grid(color='#1e293b', linestyle='--', linewidth=0.5)
ax.plot(roll_corr, color='#a78bfa', linewidth=1.2)
ax.axhline(0, color='#64748b', linewidth=0.8, linestyle='--')
ax.set_title(f'Rolling {win}-period Correlation: {top_feat} vs fwd_ret_1', color='#e2e8f0')
plt.tight_layout()
plt.show()
```

## 4. Save Features to FeatureStore

```python
store = FeatureStore()
cols_to_save = ['timestamp', 'close', 'returns'] + feature_cols + target_cols + ['alpha_score']
save_cols = [c for c in cols_to_save if c in df.columns]
store.save_features(df.select(save_cols), symbol=SYMBOL, timeframe=TIMEFRAME)
print(f"✅ Features saved for {SYMBOL}/{TIMEFRAME}: {save_cols}")
```
